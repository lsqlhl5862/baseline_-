```python
def getTetCom(text):
    "获得全文实体名称和位置信息"
    ners = fool.ner(str(text))
    com_names_pos = {}
    if len(ners[0]) == 0:
        return ''
    else:
        for ner_list in ners:
            for c_ner in ner_list:
                if 'company' in c_ner[2]:
                    pos_l = [c_ner[0], c_ner[1]]
                    com_names_pos[c_ner[-1].strip()] = pos_l
                    #com_names.append(c_ner[-1].strip())
        return com_names_pos
```

```python
def getTimeNer(one_pro_mess):
    "抽取日期实体"
    time = []
    ners_res = fool.ner(str(one_pro_mess))
    for ners in ners_res:
        if len(ners) == 0:
            continue
        else:
            if 'time' in ners[0]:
                time.append(ners[0][-1])
    return time
```

```python
def stru_mess(one_pro_mess, proc_dec):
    "输入：一款理财产品的结构化信息"
    "输出：获得需要的信息"
    one_prod_dict = {}
    time_mess = getTimeNer(proc_dec)
    time_mess = ';'.join(time_mess)
    for txt in one_pro_mess:
        txt = txt.split('：')
        if len(txt) == 2:
            attr = txt[0]
            value = txt[1]
            if '万元' in value or '亿元' in value:
                one_prod_dict['认购金额'] = value
            if '受托方' in attr or '发行人' in attr:
                one_prod_dict['产品发行方名称'] = value
            if '关联关系' in attr and '产品发行方名称' not in one_prod_dict:
                company = fool.ner(value)
                if len(company[0]) == 0:
                    continue
                else:
                    if 'company' in company[0][-1]:
                        one_prod_dict['产品发行方名称'] = company[0][-1][-1]
        one_prod_dict['时间'] = time_mess
    return one_prod_dict
```
